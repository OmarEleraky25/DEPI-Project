```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train_LZdllcl.csv")
```

```python
df_2 = pd.read_csv("department.csv")
```

```python
df_2
```

```python
merged = pd.merge(df, df_2, how='left', on='employee_id')
```

```python
merged
```

```python
def perform_eda(df):
    
    print("Basic Information:")
    print("-" * 50)
    print(f"Shape of the dataset: {df.shape}")
    print(f"\nData types:\n{df.dtypes}")
    
    print("\nMissing Values:")
    print("-" * 50)
    print(df.isnull().sum())
    

    duplicates = df.duplicated().sum()
    print(f"\nNumber of duplicate rows: {duplicates}")
    
    print("\nDescriptive Statistics:")
    print("-" * 50)
    print(df.describe())
```

```python
perform_eda(merged)
```

```python
merged.dropna(subset=['education', "previous_year_rating"], inplace=True)
```

```python
merged
```

```python
features = ['no_of_trainings','age','previous_year_rating','length_of_service','KPIs_met >80%','awards_won?','avg_training_score','is_promoted']
plt.figure(figsize=(15, 7))
for i in range(0, len(features)):
    plt.subplot(1, 8, i+1)
    sns.boxplot(y=merged[features[i]],color='green',orient='v')
    plt.tight_layout()
```

```python
title = 'Distribution of Promoted Employees based on Age'

plt.figure(figsize=(10, 7))

sns.histplot(data=merged, x="age", hue="is_promoted",bins=35, palette="Pastel1", multiple="stack")

plt.title(title)
plt.xlabel("Age")  
plt.tight_layout()


plt.show()

print("The average age of the employees is " + str(round(np.mean(merged.age), 0)))
print("The median age of the employees is " + str(np.median(merged.age)))
```

```python
prob_kpi = merged.groupby(['KPIs_met >80%','is_promoted'])['employee_id'].count().reset_index()
kpi = prob_kpi.pivot_table(index='KPIs_met >80%', columns='is_promoted', 
                       values='employee_id').reset_index()
kpi.columns = ['kpi', 'not_promoted', 'promoted']
kpi['total'] = kpi['not_promoted']+kpi['promoted']
kpi['probability'] = round((kpi['promoted']/kpi['total'])*100,2)
kpi = kpi.sort_values(['probability'], ascending=False)
kpi
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.barplot(x='kpi', y='probability', data=kpi, ax=ax, color='coral', 
            order=kpi.sort_values('probability', ascending=False).kpi)


ax.set_title('Promotion by Key Performance Indicator')

ax.set_xlabel('Key Performance Indicator')
ax.set_ylabel('Probability')
ax.set_xticklabels(['More than 80%', 'Less than 80%'])

plt.tight_layout()
plt.show()
```

Employees who achieve KPIs of more than 80% have a 17.02% chance of getting a promotion


```python
rating_probs = merged.groupby(['previous_year_rating','is_promoted']).agg({
    'department': ['count']
}).reset_index()

rating_probs.columns = ['previous_year_rating', 'is_promoted', "employees"]

rating_probs = pd.pivot_table(rating_probs,
                              index = 'previous_year_rating',
                              columns = 'is_promoted',
                              values = "employees"
                             ).reset_index()

rating_probs.columns = ['previous_year_rating', 'not_promoted', 'promoted']
rating_probs['total_employees'] = rating_probs['not_promoted'] + rating_probs['promoted']
rating_probs['promotion_probs'] = (rating_probs['promoted']/(rating_probs['not_promoted']+rating_probs['promoted']))*100
rating_probs = rating_probs.sort_values('promotion_probs', ascending=False).reset_index(drop=True)
rating_probs
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.barplot(x='previous_year_rating', y='promotion_probs', data=rating_probs,
            order=rating_probs.sort_values('promotion_probs')['previous_year_rating'],
            color='coral', ax=ax)

ax.set_title('Promotion by Previous Year Rating')

ax.set_xlabel('Previous Year Rating (1 to 5)')
ax.set_ylabel('Promotion Probabilities')

plt.tight_layout()
plt.show()
```

```python
data_awards = pd.pivot_table(merged,
                    index = ['awards_won?'],
                    columns = ['is_promoted'],
                    aggfunc = {'is_promoted' : ['count']}).reset_index()
data_awards.columns = ['awards_won', 'not_promoted', 'promoted']
data_awards['total'] = data_awards['not_promoted']+data_awards['promoted']
data_awards['probability'] = round((data_awards['promoted']/data_awards['total'])*100,2)
data_awards = data_awards.sort_values(['probability'], ascending=False)
data_awards
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.barplot(x='awards_won', y='probability', data=data_awards, ax=ax, color='coral', 
            order=data_awards.sort_values('probability', ascending=False).awards_won)


ax.set_title('Promotion by Awards Won')

ax.set_xlabel('Awards Won')
ax.set_ylabel('Probability')
ax.set_xticklabels(['Awarded', 'Not Awarded'])

plt.tight_layout()
plt.show()
```
